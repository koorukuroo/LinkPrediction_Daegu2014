```python
# 라이브러리 불러오기
import networkx as nx
```

```python
# 페이스북 Gexf 불러오기
G = nx.read_gexf('facebook.gexf')
```

```python
# 네트워크 그림 그리기
nx.draw(G)
```

```python
# G.nodes() #전체 사람 리스트 보기 / 페이스북 사람(노드) 고유번호
G.nodes()[:10] # 10명만 보기
```

```python
# 사람 두 명을 임의로 집어 넣어서 prediction 계산하기
preds = nx.adamic_adar_index(G, [('100001410409245', '100000907084163')])
for u, v, p in preds:
    print u, v, p
```

### 확률이 높은 짝을 어떻게 찾을까?

```python
# 이제 한꺼번에 많은 사람 해보기
l = G.nodes()[:50]
pair = []
for i in l:
    for j in l:
        if i != j:
            pair.append((i, j))
preds = nx.adamic_adar_index(G, pair)

index = 0
order = {} # Dict에 key를 사람 번호, value를 인덱스로 집어 넣자
for u, v, p in preds:
    order[p] = index
    index += 1
```
